```python
import zipfile
import os

zip_path = "/content/archive (2).zip"
extract_path = "/content/groundwater_data"

# Unzipping the file
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)


print("Extracted files:", os.listdir(extract_path))
```

```python
import pandas as pd
import os

csv_path = "/content/groundwater_data/gwl-monthly.csv"

df = pd.read_csv(csv_path)
df.head()
```

```python
df.info()
df.columns
df.describe()
df.isnull().sum()
```

```python
# Converting dates
df['MSMT_DATE'] = pd.to_datetime(df['MSMT_DATE'])

# Sorting by station and date
df = df.sort_values(by=['STATION', 'MSMT_DATE'])

# Dropping columns with too many nulls
df_clean = df.drop(columns=['GSE_WSE', 'RPE_WSE', 'WSE'])

# Dropping rows missing WLM_RPE
df_clean = df_clean.dropna(subset=['WLM_RPE'])

# Cleaned data
df_clean.head()
```

```python
df_clean.to_csv("/content/cleaned_groundwater.csv", index=False)
```
